```python
# Check for and remove duplicate entries in the products dataframe
products_df = pd.read_csv("/wk8-products.csv")
initial_products_rows = len(products_df)
print(f"Initial number of rows in products dataframe: {initial_products_rows}")
products_duplicates_found = products_df.duplicated().sum()
if products_duplicates_found > 0:
    products_df.drop_duplicates(inplace=True)
    print(f"Number of duplicate rows removed from products dataframe: {products_duplicates_found}")
    print(f"New number of rows in products dataframe: {len(products_df)}")
else:
    print("No duplicate rows found in products dataframe.")

print("-" * 30)

# Check for and remove duplicate entries in the customers dataframe
customers_df = pd.read_csv("/wk8-customers.csv")
initial_customers_rows = len(customers_df)
print(f"Initial number of rows in customers dataframe: {initial_customers_rows}")
customers_duplicates_found = customers_df.duplicated().sum()
if customers_duplicates_found > 0:
    customers_df.drop_duplicates(inplace=True)
    print(f"Number of duplicate rows removed from customers dataframe: {customers_duplicates_found}")
    print(f"New number of rows in customers dataframe: {len(customers_df)}")
else:
    print("No duplicate rows found in customers dataframe.")
```
